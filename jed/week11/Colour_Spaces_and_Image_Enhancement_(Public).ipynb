```python
#@title Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#@title Load the medical image and convert to grayscale and HSV color spaces
image = cv2.imread("./img/CT_image.png")
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
#@title Apply histogram equalization
equalized_image = cv2.equalizeHist(grayscale_image)
```

```python
#@title Apply CLAHE
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_image = clahe.apply(grayscale_image)
```

```python
#@title Contrast stretching
stretched_image = cv2.normalize(grayscale_image, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
```

```python
#@title Apply Canny edge detection
edges = cv2.Canny(grayscale_image, 100, 200)
```

```python
#@title Display the images
fig, axes = plt.subplots(2, 3, figsize=(15, 9))
axes[0, 0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title("Original Image")
axes[0, 1].imshow(grayscale_image, cmap='gray')
axes[0, 1].set_title("Grayscale Image")
axes[0, 2].imshow(hsv_image, cmap='hsv')
axes[0, 2].set_title("HSV Image")
axes[1, 0].imshow(equalized_image, cmap='gray')
axes[1, 0].set_title("Histogram Equalized Image")
axes[1, 1].imshow(clahe_image, cmap='gray')
axes[1, 1].set_title("CLAHE Image")
axes[1, 2].imshow(edges, cmap='gray')
axes[1, 2].set_title("Canny Edge Detection")

for ax in axes.flatten():
    ax.axis("off")

plt.show()
```
